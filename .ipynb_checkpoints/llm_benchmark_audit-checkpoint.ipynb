# Are Open LLM Leaderboard Rankings Real?
### A statistical audit of 4,576 model rankings

The Open LLM Leaderboard ranks models using an average of six benchmark scores and presents
them as an ordered list — #1, #2, #3, ... #4576. This notebook asks a simple question:
**how many of those ranks are statistically distinguishable from each other?**

We also ask whether the six benchmarks are actually measuring six different things,
or whether some are so correlated they're just double-counting the same signal.

**The answer is uncomfortable.** With honest statistics, 4,576 ranked models collapse
into just 22 statistically distinguishable groups but the leaderboard presents 4,576
significant figures.

## 2. Clean and Select Relevant Columns

We keep the six benchmark **Raw** scores (actual proportions, 0–1 scale),
the overall average, model name, and parameter count.

Raw scores are what we need for CI math as they represent the actual fraction
of questions answered correctly, before any normalization.


````python
print(df_raw[["GPQA Raw", "MMLU-PRO Raw", "BBH Raw"]].describe())
````

````python
BENCH_RAW = {
    "GPQA":     "GPQA Raw",
    "MMLU_PRO": "MMLU-PRO Raw",
    "MATH":     "MATH Lvl 5 Raw",
    "BBH":      "BBH Raw",
    "IFEval":   "IFEval Raw",
    "MUSR":     "MUSR Raw",
}

# Question counts per benchmark (fixed, from official documentation)
BENCH_N = {
    "GPQA":     448,    # 448 graduate-level questions
    "MMLU_PRO": 12032,  # 12,032 questions across 14 domains
    "MATH":     1324,   # Level 5 competition math subset
    "BBH":      6511,   # 23 subtasks, ~250-500 each
    "IFEval":   541,    # 541 prompts with verifiable instructions
    "MUSR":     756,    # 3 subtasks: murder mystery (250), object placement (256), team allocation (250)
}

cols_keep = {
    "fullname": "model",
    "#Params (B)": "params_b",
    "Average ⬆️": "avg_score",
    **{v: k for k, v in BENCH_RAW.items()}
}

df = df_raw[list(cols_keep.keys())].rename(columns=cols_keep).copy()

for bench in BENCH_RAW.keys():
    df[bench] = pd.to_numeric(df[bench], errors="coerce")

df["avg_score"] = pd.to_numeric(df["avg_score"], errors="coerce") / 100.0
df["params_b"]  = pd.to_numeric(df["params_b"],  errors="coerce")

df = df.dropna(subset=list(BENCH_RAW.keys()), how="all").reset_index(drop=True)

# Original rank = position in the leaderboard (sorted by avg_score descending)
df = df.sort_values("avg_score", ascending=False).reset_index(drop=True)
df["orig_rank"] = df.index + 1

print(f"  Benchmarks: {list(BENCH_RAW.keys())}")
print(f"\nScore ranges:")
for b in BENCH_RAW.keys():
    print(f"  {b:10s}: {df[b].min():.3f} – {df[b].max():.3f}  (n questions = {BENCH_N[b]:,})")
````

---
## 3. Compute Confidence Intervals

### Why do we need CIs?

When a model scores 73.4% on a benchmark with 448 questions, that's an estimate. If the same model sat the benchmark again with a different random sample
of questions from the same distribution, it wouldn't score exactly 73.4% again.

The confidence interval tells us the plausible range for the model's true score.
**Without it, we can't know if two models are genuinely different or just noise.**

### Method per benchmark

| Benchmark | Method | Why |
|-----------|--------|-----|
| GPQA | Proportion CI (Wilson) | Right/wrong, single proportion |
| MMLU-PRO | Proportion CI (Wilson) | Right/wrong, single proportion |
| MATH Lvl 5 | Proportion CI (Wilson) | Right/wrong, single proportion |
| BBH | Bootstrap | Averages across 23 subtasks of unequal size |
| IFEval | Bootstrap | Two-level nesting (prompts → instructions) |
| MUSR | Bootstrap | Subtask averaging + narrative correlation |

The **Wilson score interval** is more accurate than the textbook formula when
scores are near 0 or 1. For bootstrap, we resample 2,000 times.


````python
from scipy.stats import norm

Z = 1.96  # 95% confidence level

def wilson_ci(p, n):
    """Wilson score confidence interval for a proportion p with sample size n."""
    if pd.isna(p) or pd.isna(n) or n == 0:
        return (np.nan, np.nan)
    p = np.clip(p, 0, 1)
    denom = 1 + Z**2 / n
    centre = (p + Z**2 / (2*n)) / denom
    margin = (Z / denom) * np.sqrt(p*(1-p)/n + Z**2/(4*n**2))
    return (max(0, centre - margin), min(1, centre + margin))

def bootstrap_ci(score, n, n_boot=2000, seed=42):
    """
    Bootstrap CI: simulate n_boot samples of size n with the observed proportion,
    return the 2.5th and 97.5th percentiles.
    """
    if pd.isna(score) or pd.isna(n) or n == 0:
        return (np.nan, np.nan)
    rng = np.random.default_rng(seed)
    p = np.clip(score, 0, 1)
    samples = rng.binomial(n, p, size=n_boot) / n
    return (np.percentile(samples, 2.5), np.percentile(samples, 97.5))

# Which benchmarks use which method
PROP_BENCHES      = ["GPQA", "MMLU_PRO", "MATH"]
BOOTSTRAP_BENCHES = ["BBH", "IFEval", "MUSR"]

for bench in PROP_BENCHES:
    n = BENCH_N[bench]
    cis = df[bench].apply(lambda p: wilson_ci(p, n))
    df[f"{bench}_lo"] = cis.apply(lambda x: x[0])
    df[f"{bench}_hi"] = cis.apply(lambda x: x[1])
    df[f"{bench}_width"] = df[f"{bench}_hi"] - df[f"{bench}_lo"]
    print(f"  ✓ {bench} (Wilson, n={n:,})")

for bench in BOOTSTRAP_BENCHES:
    n = BENCH_N[bench]
    cis = df[bench].apply(lambda p: bootstrap_ci(p, n))
    df[f"{bench}_lo"] = cis.apply(lambda x: x[0])
    df[f"{bench}_hi"] = cis.apply(lambda x: x[1])
    df[f"{bench}_width"] = df[f"{bench}_hi"] - df[f"{bench}_lo"]
    print(f"  ✓ {bench} (Bootstrap, n={n:,})")

print("\n✓ Done. CI width summary (percentage points):")
for bench in list(BENCH_RAW.keys()):
    w = df[f"{bench}_width"].dropna() * 100
    print(f"  {bench:10s}: median ±{w.median()/2:.1f}pp  |  max ±{w.max()/2:.1f}pp")
````

---
## 4. Propagate Uncertainty to the Average Score

The leaderboard average is the mean of six benchmark scores.
When each benchmark has its own uncertainty, the average does too.

### The correct way to do this

Each benchmark gives us a standard error (SE) for that score.
The SE of an average of k independent measurements is:

```
SE_avg = sqrt( sum(SE_i²) ) / k
```

This is smaller than any individual SE — averaging reduces noise.
The 95% CI on the average is then: `avg_score ± 1.96 * SE_avg`


````python
all_benches = list(BENCH_RAW.keys())
k = len(all_benches)  # number of benchmarks = 6

# Step 1: derive SE for each benchmark from its CI
# CI width = 2 * 1.96 * SE  →  SE = width / (2 * 1.96)
for b in all_benches:
    df[f"{b}_se"] = df[f"{b}_width"] / (2 * 1.96)

# Step 2: SE of the average = sqrt(sum of SE²) / k
se_sq_cols = [f"{b}_se" for b in all_benches]
df["avg_se"] = np.sqrt(df[se_sq_cols].pow(2).sum(axis=1)) / k

# Step 3: 95% CI on the average
df["avg_lo"]    = df["avg_score"] - 1.96 * df["avg_se"]
df["avg_hi"]    = df["avg_score"] + 1.96 * df["avg_se"]
df["avg_width"] = df["avg_hi"] - df["avg_lo"]

print("Combined average CI half-width (percentage points):")
w = df["avg_width"].dropna() * 100
print(f"  Median: ±{w.median()/2:.2f}pp")
print(f"  Mean:   ±{w.mean()/2:.2f}pp")
print(f"  Max:    ±{w.max()/2:.2f}pp")
print(f"\n  (Correctly narrower than any single benchmark — averaging reduces noise)")

print(f"\nTop 20 models — score with 95% CI:")
top20 = df.head(20)[["model","avg_score","avg_lo","avg_hi"]].copy()
top20["score_pct"]   = (top20["avg_score"]*100).round(2)
top20["ci_lo_pct"]   = (top20["avg_lo"]*100).round(2)
top20["ci_hi_pct"]   = (top20["avg_hi"]*100).round(2)
top20["ci_display"]  = top20.apply(
    lambda r: f"{r.score_pct:.1f}%  [{r.ci_lo_pct:.1f}, {r.ci_hi_pct:.1f}]", axis=1)
top20["model_short"] = top20["model"].str.split("/").str[-1].str[:40]
print(top20[["model_short","ci_display"]].to_string(index=False))
````

````python
### What ±1.18pp actually means

The combined CI half-width across all models is approximately ±1.18 percentage
points. This means two models need to differ by more than ~2.4pp for us to
reliably tell them apart at 95% confidence.

The entire spread of scores in the top 100 models runs from ~42% to ~52%
a range of 10 percentage points. With a detection threshold of 2.4pp, only
gaps larger than roughly a quarter of that total range are statistically real.
Everything closer than that is noise the leaderboard is presenting as signal.
````

---
## 5. Collapse Rankings into Real Statistical Tiers

**The core finding.**

A model joins the current tier only if its CI overlaps with the **anchor** which is
the highest-scoring model that started that tier. We compare each new model's
CI upper bound against the anchor's CI lower bound. If they don't overlap,
that model is genuinely distinguishable from the best in the tier and starts
a new tier.

**What the result means:**
The leaderboard presents every model as uniquely ranked. Our tiers show
how many of those ranks represent genuinely different capability levels,
and how many are positions within the same statistical blur.

````python
def assign_tiers(df_sorted):
    """
    Walk the ranked list top to bottom.

    A model joins the current tier only if its CI overlaps with the ANCHOR
    the highest-scoring model that started this tier
    """
    tiers = []
    tier_id = 1
    anchor_lo = df_sorted.iloc[0]["avg_lo"]  # lower bound of the tier's top model

    for _, row in df_sorted.iterrows():
        if row["avg_hi"] < anchor_lo:        # can't reach the anchor so new tier
            tier_id += 1
            anchor_lo = row["avg_lo"]        # this model is now the new anchor
        tiers.append(tier_id)

    return tiers

# Only use models with complete CI data
df_valid = df.dropna(subset=["avg_lo","avg_hi"]).copy()
df_valid = df_valid.sort_values("avg_score", ascending=False).reset_index(drop=True)
df_valid["tier"] = assign_tiers(df_valid)

n_tiers  = df_valid["tier"].max()
n_models = len(df_valid)

print(f"Models analysed:       {n_models:,}")
print(f"Original ranks:        {n_models:,}  (every model gets a unique rank)")
print(f"Statistical tiers:     {n_tiers:,}  (groups of truly distinguishable models)")
print(f"Compression ratio:     {n_models/n_tiers:.0f}x  (each 'real rank' covers ~{n_models/n_tiers:.0f} models on average)")

tier_sizes = df_valid["tier"].value_counts().sort_index()
print(f"\nTier size distribution:")
print(f"  Median models per tier: {tier_sizes.median():.0f}")
print(f"  Largest tier:           {tier_sizes.max()} models (statistically identical)")
print(f"  Tiers with >10 models:  {(tier_sizes > 10).sum()}")
````

## 6. Visualisations

### 6a. CI width by benchmark

How uncertain are we about each benchmark's score? This matters because
a benchmark with wide CIs contributes more noise to the average than one with tight CIs.


````python
fig, ax = plt.subplots(figsize=(9, 4))

bench_labels = list(BENCH_RAW.keys())

# Compute CI width at p=0.5 (worst case / most honest comparison)
# This removes the distortion from thousands of near-zero scoring models dragging down the median.
ref_widths = []
for b in bench_labels:
    n = BENCH_N[b]
    if b in PROP_BENCHES:
        lo, hi = wilson_ci(0.5, n)
    else:
        lo, hi = bootstrap_ci(0.5, n)
    ref_widths.append((hi - lo) * 100)

colors = ["#D85A30" if b in ["MUSR", "GPQA"] else "#1D9E75" for b in bench_labels]
bars = ax.barh(bench_labels, ref_widths, color=colors, height=0.55, edgecolor="white")

for bar, w in zip(bars, ref_widths):
    ax.text(w + 0.02, bar.get_y() + bar.get_height()/2,
            f"±{w/2:.2f}pp", va="center", fontsize=10)

ax.set_xlabel("95% CI width at p=0.5 (percentage points)")
ax.set_title("Benchmark precision: how wide is the CI at a 50% score?\n"
             "Wider = more noise contributed to the average (narrower = more precise)", fontsize=11)

green_patch = mpatches.Patch(color="#1D9E75", label="Reliable CI")
red_patch   = mpatches.Patch(color="#D85A30", label="Wide CI")
ax.legend(handles=[green_patch, red_patch], fontsize=9)

plt.tight_layout()
plt.savefig("fig1_ci_width.png", dpi=150, bbox_inches="tight")
plt.show()
````

### 6b. Top 50 models: original rank vs CI range

This is the headline chart. Each horizontal bar shows a model's 95% CI range.
Overlapping bars belong to the same statistical tier — their ranking is noise.


````python
top_n = 50
df_top = df_valid.head(top_n).copy()
df_top["model_short"] = df_top["model"].str.split("/").str[-1].str[:35]

fig, ax = plt.subplots(figsize=(11, 14))

# Color bars by tier
palette = plt.cm.tab20.colors
tier_colors = {t: palette[i % 20] for i, t in enumerate(df_top["tier"].unique())}

for i, (_, row) in enumerate(df_top.iterrows()):
    y = top_n - i
    color = tier_colors.get(row["tier"], "steelblue")
    ax.barh(y, row["avg_hi"] - row["avg_lo"],
            left=row["avg_lo"], height=0.7,
            color=color, alpha=0.75, edgecolor="white", linewidth=0.5)
    ax.plot(row["avg_score"], y, "o", color="black", markersize=3, zorder=5)

ax.set_yticks(range(top_n, 0, -1))
ax.set_yticklabels(df_top["model_short"].tolist(), fontsize=7.5)
ax.set_xlabel("Average benchmark score (95% CI)", fontsize=11)
ax.set_title(
    f"Top {top_n} models: point estimate (dot) + 95% confidence interval (bar)\n"
    "Same color = same statistical tier = cannot be ranked apart", fontsize=11)

plt.tight_layout()
plt.savefig("fig2_top50_ci.png", dpi=150, bbox_inches="tight")
plt.show()
````

````python
Each horizontal bar shows a model's 95% CI on its combined average score.
Models sharing the same color belong to the same statistical tier — they
cannot be ranked against each other at 95% confidence.

The clear visual gap between the top 4 (around 0.50–0.52) and the next
group (around 0.46–0.48) represents a real capability difference of roughly
3–4 percentage points — detectable at our ±1.18pp threshold. Within each
color band however, the leaderboard's rankings are noise.
````

### 6c. How many real tiers are there?

The distribution of tier sizes tells us how badly the leaderboard is
over-stating its own precision.


````python
tier_sizes = df_valid["tier"].value_counts().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: histogram of tier sizes
axes[0].hist(tier_sizes.values, bins=40, color="#378ADD", edgecolor="white", alpha=0.85)
axes[0].set_xlabel("Number of models in tier")
axes[0].set_ylabel("Number of tiers")
axes[0].set_title("Tier size distribution\n(how many models per 'real rank')")
axes[0].axvline(tier_sizes.median(), color="red", linestyle="--",
                label=f"Median = {tier_sizes.median():.0f}")
axes[0].legend()

# Right: cumulative — what % of tiers account for what % of models?
cumulative_models = tier_sizes.sort_values(ascending=False).cumsum() / n_models * 100
axes[1].plot(range(1, len(cumulative_models)+1), cumulative_models.values,
             color="#1D9E75", linewidth=2)
axes[1].set_xlabel("Number of largest tiers")
axes[1].set_ylabel("% of all models covered")
axes[1].set_title("Cumulative model coverage\nby tier size (largest first)")
axes[1].axhline(50, color="gray", linestyle="--", alpha=0.5, label="50% of models")
axes[1].legend()

fig.suptitle(
    f"{n_models:,} models → {n_tiers:,} real tiers  "
    f"({n_models/n_tiers:.0f}x compression)", fontsize=12, y=1.02)

plt.tight_layout()
plt.savefig("fig3_tier_distribution.png", dpi=150, bbox_inches="tight")
plt.show()
````

## 7. Are the Six Benchmarks Measuring Different Things?

### The redundancy problem

If two benchmarks are highly correlated, equal-weighting them in the average
is like counting the same thing twice. We check:

1. The overall correlation matrix across all 4,576 models
2. Whether the correlation **changes within size tiers** — because our earlier
   analysis showed the 0.96 BBH/MMLU-PRO correlation might be a scale artifact


````python
bench_list = list(BENCH_RAW.keys())
bench_data = df_valid[bench_list].dropna()

corr = bench_data.corr(method="spearman")

fig, ax = plt.subplots(figsize=(7, 6))
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="RdYlGn",
            vmin=0, vmax=1, ax=ax, square=True,
            linewidths=0.5, cbar_kws={"shrink": 0.8},
            mask=False)
ax.set_title("Spearman correlation between benchmarks\n(across all models)", fontsize=11)
plt.tight_layout()
plt.savefig("fig4_correlation_all.png", dpi=150, bbox_inches="tight")
plt.show()

print("\nHighest correlations (potential redundancy):")
corr_pairs = []
for i, b1 in enumerate(bench_list):
    for j, b2 in enumerate(bench_list):
        if j > i:
            corr_pairs.append((b1, b2, corr.loc[b1, b2]))
corr_pairs.sort(key=lambda x: abs(x[2]), reverse=True)
for b1, b2, r in corr_pairs[:5]:
    flag = " ← REDUNDANT" if abs(r) > 0.9 else ""
    print(f"  {b1:10s} × {b2:10s}: r = {r:.3f}{flag}")
````

### 7b. Does correlation change within model size tiers?

The 0.96 correlation between BBH and MMLU-PRO across all models is likely driven
by scale — big models score high on both, small models score low on both.

If we split by model size and recompute, does the correlation collapse?
**If yes, the benchmarks actually do measure different things at the frontier
— exactly where ranking precision matters most.**


````python
# Define size buckets
df_valid["size_tier"] = pd.cut(
    df_valid["params_b"],
    bins=[0, 3, 10, 30, 80, np.inf],
    labels=["<3B", "3–10B", "10–30B", "30–80B", ">80B"]
)

# Focus on BBH vs MMLU_PRO (the most correlated pair)
pair = ("BBH", "MMLU_PRO")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left: scatter all models
axes[0].scatter(df_valid[pair[0]], df_valid[pair[1]],
                alpha=0.15, s=8, color="#378ADD")
r_all = df_valid[[pair[0], pair[1]]].dropna().corr(method="spearman").iloc[0,1]
axes[0].set_title(f"All models  (r = {r_all:.3f})", fontsize=11)
axes[0].set_xlabel(pair[0]); axes[0].set_ylabel(pair[1])

# Right: correlation by size tier
size_labels = ["<3B", "3–10B", "10–30B", "30–80B", ">80B"]
corrs, counts = [], []
for label in size_labels:
    sub = df_valid[df_valid["size_tier"] == label][[pair[0], pair[1]]].dropna()
    if len(sub) >= 10:
        r = sub.corr(method="spearman").iloc[0, 1]
        corrs.append(r)
        counts.append(len(sub))
    else:
        corrs.append(np.nan)
        counts.append(0)

bar_colors = ["#D85A30" if r > 0.85 else "#1D9E75" for r in corrs]
bars = axes[1].bar(size_labels, corrs, color=bar_colors,
                   edgecolor="white", alpha=0.85)
axes[1].axhline(0.9, color="red", linestyle="--", alpha=0.6, label="r = 0.9 threshold")
axes[1].set_ylim(0, 1.05)
axes[1].set_ylabel("Spearman r")
axes[1].set_title(f"BBH × MMLU-PRO correlation\nbroken down by model size", fontsize=11)
axes[1].legend(fontsize=9)

for bar, r, n in zip(bars, corrs, counts):
    if not np.isnan(r):
        axes[1].text(bar.get_x() + bar.get_width()/2, r + 0.01,
                     f"r={r:.2f}\nn={n}", ha="center", fontsize=8.5)

plt.tight_layout()
plt.savefig("fig5_correlation_by_size.png", dpi=150, bbox_inches="tight")
plt.show()

print("\nCorrelation within size buckets:")
for label, r, n in zip(size_labels, corrs, counts):
    if not np.isnan(r):
        flag = " ← correlation drops here" if r < 0.85 else ""
        print(f"  {label:8s}: r = {r:.3f}  (n={n}){flag}")
````

## 9. Conclusions

### What we found

**Finding 1 — The leaderboard has 4,576 ranks. Statistically, it has 22.**
With calculated confidence intervals and a precision of ±1.18 percentage points
on the combined average score, 4,576 models collapse into just 22 statistically
distinguishable groups. The typical leaderboard "rank" is actually a group of
~208 statistically identical models. The largest single tier contains 515 models
the leaderboard presents as having distinct positions.

**Finding 2 — BBH and MMLU-PRO are nearly redundant at population level.**
These two benchmarks correlate at 0.95 across all models, meaning equal-weighting
them double-counts the same signal. Critically, this correlation is driven by
scale as bigger models score higher on both. Within the >80B frontier tier the
correlation drops to 0.76, suggesting the benchmarks do measure different things
at the level where ranking precision matters most, but the leaderboard's equal
weighting doesn't account for this.

**Finding 3 — The precision problem is structural, not fixable by dropping benchmarks.**
Removing MUSR does not improve the number of distinguishable tiers. The bottleneck
is question count across the board: GPQA (448 questions) and IFEval (541 prompts)
each contribute ±4+ percentage point CIs. The combined average can only achieve
±1.18pp precision, which is too wide to distinguish models that differ by less
than ~2.4pp — the majority of models in the top 100.

### What this means

The field routinely cites leaderboard positions to justify architectural choices,
training decisions, and deployment recommendations. This analysis shows that a
large fraction of those position differences are below the threshold of statistical
significance. We are reading 4,576 significant figures from a ruler with 22 tick marks.

### What should change

1. **Leaderboards should display CIs**, not just point estimates. A model ranked
   #47 that is statistically identical to #112 should be shown as such.
2. **GPQA needs more questions.** At 448 questions it is the precision bottleneck
   despite not being the smallest benchmark by count.
3. **Equal weighting needs justification.** BBH and MMLU-PRO are 0.95 correlated
   — weighting them equally double-counts one signal while other dimensions
   (instruction following via IFEval, narrative reasoning via MUSR) get equal billing
   despite measuring genuinely distinct things.
4. **Correlation structure should inform weighting.** At the frontier (>80B models)
   the BBH/MMLU-PRO correlation drops to 0.76 — suggesting differentiated weighting
   at different capability levels would produce more informative rankings.
